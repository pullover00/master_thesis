```python
file_name ='/content/YOLOX/YOLOX_outputs/yolox_voc_s/train_log.txt'
  
# opening and reading the file 
file = open(file_name, 'r')
lines = file.readlines()

new_list = []
text = 'iter: 60/60'
loss = 'total_loss'
for line in lines:
        if text in line:
          index = line.find(loss)
          epoch_loss = float(line[index + 12:index + 15])
          new_list.append(epoch_loss)
          if len(new_list) == 300:
            break
print(new_list)

```

```python
file_name ='/content/YOLOX/YOLOX_outputs/yolox_voc_s/val_log.txt'
  
# opening and reading the file 
file = open(file_name, 'r')
lines = file.readlines()

new_list = []
text = 'iter: 60/60'
loss = 'total_loss'
for line in lines:
        if text in line:
          index = line.find(loss)
          epoch_loss = float(line[index + 12:index + 15])
          new_list.append(epoch_loss)
          if len(new_list) == 300:
            break
print(new_list)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import os 
import numpy as np

epoch_nums = 300
x = np.arange(0, epoch_nums, 1)

# calculate trend line #calculate equation for trendline
z = np.polyfit(x, new_list, 2)
p = np.poly1d(z)

#add trendline to plot
plt.plot(x, p(x))

plt.rc('font', family='Times New Roman', size=13)  #  The global Chinese and English fonts are “ roman type ”
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

 #########################################  Corresponding to oneself epoch
plt.plot(x, new_list)
plt.plot(x, p(x))

plt.xlabel("Epoch")
plt.ylabel("total loss")
plt.xlim(0, epoch_nums) ####################################################  ditto  


#x_major_locator = MultipleLocator(10)  #  hold x The scale interval of the axis is set to 10, And exist in variables  ###############################  Set the axis spacing 
#ax = plt.gca()
#ax.xaxis.set_major_locator(x_major_locator)

plt.grid(True)
plt.show()
```
